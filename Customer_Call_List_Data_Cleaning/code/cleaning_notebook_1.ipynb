```python
# importing the required modules
import pandas as pd
import numpy as np
```

##### Reading in the dataset.

```python
call_list = pd.read_excel("D:\\Data_Analysis\\Data_Cleaning\\Customer_Call_List_Data_Cleaning\\dataset\\Customer Call List.xlsx")

call_list
```

##### Dropping duplicate rows.

```python
call_list.drop_duplicates(inplace = True)
call_list
```

```python
call_list.info()
```

##### Cleaning extra characters (by stripping) in the column[Last_Name].

```python
call_list["Last_Name"] = call_list["Last_Name"].str.strip("./_")
```

```python
call_list
```

##### Formatting the phone_number column.

```python
call_list["Phone_Number"] = call_list["Phone_Number"].str.replace("[^0-9]", "", regex = True)
```

```python
call_list
```

```python
call_list.info()
```

```python
call_list["Phone_Number"] = call_list["Phone_Number"].apply(lambda x: f"{x[0:3]}-{x[3:6]}-{x[6:]}" if len(str(x)) == 10 else np.nan)
```

```python
call_list
```

##### Cleaning the address column.

```python
extra_add_data = call_list["Address"].str.split(", ").str[1] # getting data for the new column
call_list.insert(5, "Extra_Address_Info", extra_add_data)

call_list["Address"] = call_list["Address"].str.split(", ").str[0]
```

```python
call_list["Extra_Address_Info"].fillna("No Extra Information", inplace = True)
```

```python
call_list
```

Replacing some unusual data with usual data.

```python
call_list["Paying Customer"].replace({"Y" : "Yes", "N" : "No"}, inplace = True)
call_list["Do_Not_Contact"].replace({"Y" : "Yes", "N" : "No"}, inplace = True)
```

```python
call_list
```

##### Deleting the not useful column.

```python
call_list.drop(columns = ["Not_Useful_Column"], inplace = True)
```

```python
call_list
```

##### Final dropping all rows of customers who have no contact number.

```python
call_list.dropna(subset = ["Phone_Number"], how = "all", inplace = True)
call_list.reset_index(drop = True, inplace = True)
```

```python
call_list.info()
```

```python
call_list
```

```python
call_list.isna().sum()
```

##### Saving the cleaned dataset.

```python
call_list.to_csv("D:\\Data_Analysis\\Data_Cleaning\\Customer_Call_List_Data_Cleaning\\dataset\\Customer Call List Cleaned.csv", index = False)
```
